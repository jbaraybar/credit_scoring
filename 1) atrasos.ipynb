```python
def update_vals_cero(x):
    if x == 0:
        return 'CERO'
    else:
        return 'NOCERO'

def update_vals_negativo(x):
    if (x <= 0):
        return 0
    else:
        return x

def valor_maximo(row):
    if row['atr_360_dias_a_mas'] > 0: return row['atr_360_dias_a_mas']
    elif row['atr_181_360_dias'] > 0: return row['atr_181_360_dias']
    elif row['atr_151_180_dias'] > 0: return row['atr_151_180_dias']
    elif row['atr_121_150_dias'] > 0: return row['atr_121_150_dias']
    elif row['atr_91_120_dias'] > 0: return row['atr_91_120_dias']
    elif row['atr_61_90_dias'] > 0: return row['atr_61_90_dias']
    elif row['atr_46_60_dias'] > 0: return row['atr_46_60_dias']
    elif row['atr_31_45_dias'] > 0: return row['atr_31_45_dias']
    elif row['atr_16_30_dias'] > 0: return row['atr_16_30_dias']
    elif row['atr_8_15_dias'] > 0: return row['atr_8_15_dias']
    elif row['atr_1_7_dias'] > 0: return row['atr_1_7_dias']
    elif row['en_plazo'] > 0: return row['en_plazo']
    else: return 0
    
def valor_maximo_posicion(row):
    if row['atr_360_dias_a_mas'] > 0: return 11
    elif row['atr_181_360_dias'] > 0: return 10
    elif row['atr_151_180_dias'] > 0: return 9
    elif row['atr_121_150_dias'] > 0: return 8
    elif row['atr_91_120_dias'] > 0: return 7
    elif row['atr_61_90_dias'] > 0: return 6
    elif row['atr_46_60_dias'] > 0: return 5
    elif row['atr_31_45_dias'] > 0: return 4
    elif row['atr_16_30_dias'] > 0: return 3
    elif row['atr_8_15_dias'] > 0: return 2
    elif row['atr_1_7_dias'] > 0: return 1
    elif row['en_plazo'] > 0: return 0
    else: return -1
    
def update_saldo_maximo(x):
    global valor
    if x['posicion'] >= 0:
        valor = x['saldomax_moratmp']
        if x['en_plazo'] < 0: valor = valor + x['en_plazo']
        if x['atr_1_7_dias'] < 0: valor = valor + x['atr_1_7_dias']
        if x['atr_8_15_dias'] < 0: valor = valor + x['atr_8_15_dias']
        if x['atr_16_30_dias'] < 0: valor = valor + x['atr_16_30_dias']
        if x['atr_31_45_dias'] < 0: valor = valor + x['atr_31_45_dias']
        if x['atr_46_60_dias'] < 0: valor = valor + x['atr_46_60_dias']
        if x['atr_61_90_dias'] < 0: valor = valor + x['atr_61_90_dias']
        if x['atr_91_120_dias'] < 0: valor = valor + x['atr_91_120_dias']
        if x['atr_121_150_dias'] < 0: valor = valor + x['atr_121_150_dias']
        if x['atr_151_180_dias'] < 0: valor = valor + x['atr_151_180_dias']
        if x['atr_181_360_dias'] < 0: valor = valor + x['atr_181_360_dias']
        if x['atr_360_dias_a_mas'] < 0: valor = valor + x['atr_360_dias_a_mas']
        return valor
    else:
        return 0

def tramo_mora(x):
    if x['saldo'] == 0 : return 0
    elif x['atr_360_dias_a_mas'] / x['saldo'] > 0.2 : return 11
    elif (x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) /x['saldo'] > 0.2 : return 10
    elif (x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 9
    elif (x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 8
    elif (x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 7
    elif (x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 6
    elif (x['atr_46_60_dias'] + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 5
    elif (x['atr_31_45_dias'] + x['atr_46_60_dias'] + + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 4
    elif (x['atr_16_30_dias'] + x['atr_31_45_dias'] + x['atr_46_60_dias'] + + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 3
    elif (x['atr_8_15_dias'] + x['atr_16_30_dias'] + x['atr_31_45_dias'] + x['atr_46_60_dias'] + + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 2
    elif (x['atr_1_7_dias'] + x['atr_8_15_dias'] + x['atr_16_30_dias'] + x['atr_31_45_dias'] + x['atr_46_60_dias'] + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo'] > 0.2 : return 1
    else: return 0

def deuda_ult_seis_meses(x):
    global acum
    acum = 0
    if (x['deuda_tot_01_A'] > 0): acum = acum + 1
    if (x['deuda_tot_02_A'] > 0): acum = acum + 1
    if (x['deuda_tot_03_A'] > 0): acum = acum + 1
    if (x['deuda_tot_04_A'] > 0): acum = acum + 1
    if (x['deuda_tot_05_A'] > 0): acum = acum + 1
    if (x['deuda_tot_06_A'] > 0): acum = acum + 1
    return acum

def estado_mora_actual(x):
    if (x['nuevo_tramo'] == 0) | (pd.isnull(x['nuevo_tramo'])):
        return '1. Al dia'
    else:
        return '2. En mora'

def estado_mora_anterior(x):
    global var_temp
    var_temp = 0
    for i in range(1, 7):
        if pd.notna(x['tramo_mora_princip_0' + str(i) + '_A']):
            var_temp = var_temp + x['tramo_mora_princip_0' + str(i) + '_A']

    if ((var_temp == 0) | (pd.isnull(var_temp))) & (x['num_meses_con_deuda_ult6'] > 0) :
        return '1. Siempre al dia'
    elif (var_temp > 0) & (x['num_meses_con_deuda_ult6'] > 0) :
        return '2. Alguna vez en mora'
    else:
        return 'Sin Informacion'

def seleccionar_segmento(x):
    if (x['est_mora_act'] == '1. Al dia') & (x['est_mora_ant_6m'] == '1. Siempre al dia') : return 'A'
    elif (x['est_mora_act'] == '2. En mora') & (x['est_mora_ant_6m'] == '1. Siempre al dia') : return 'B'
    elif (x['est_mora_act'] == '1. Al dia') & (x['est_mora_ant_6m'] == '2. Alguna vez en mora') : return 'C'
    elif (x['est_mora_act'] == '2. En mora') & (x['est_mora_ant_6m'] == '2. Alguna vez en mora') : return 'D'
    elif (x['est_mora_act'] == '1. Al dia') & (x['est_mora_ant_6m'] == 'Sin Informacion') : return 'E'
    else : return 'F'

def ratio_mora_a_tres_meses(x):
    global valor_tres_meses
    valor_tres_meses = 0

    if x['tramo_mora_princip_01_A'] > 2: valor_tres_meses = valor_tres_meses + 1
    if x['tramo_mora_princip_02_A'] > 2: valor_tres_meses = valor_tres_meses + 1
    if x['tramo_mora_princip_03_A'] > 2: valor_tres_meses = valor_tres_meses + 1
    if x['tramo_mora_princip_04_A'] > 2: valor_tres_meses = valor_tres_meses + 1
    if x['tramo_mora_princip_05_A'] > 2: valor_tres_meses = valor_tres_meses + 1
    if x['tramo_mora_princip_06_A'] > 2: valor_tres_meses = valor_tres_meses + 1
    if x['num_meses_con_deuda_ult6'] == 0: valor_tres_meses = -1

    return valor_tres_meses

def ratio_mora_a_cinco_meses(x):
    global valor_cinco_meses
    valor_cinco_meses = 0

    if x['tramo_mora_princip_01_A'] > 4: valor_cinco_meses = valor_cinco_meses + 1
    if x['tramo_mora_princip_02_A'] > 4: valor_cinco_meses = valor_cinco_meses + 1
    if x['tramo_mora_princip_03_A'] > 4: valor_cinco_meses = valor_cinco_meses + 1
    if x['tramo_mora_princip_04_A'] > 4: valor_cinco_meses = valor_cinco_meses + 1
    if x['tramo_mora_princip_05_A'] > 4: valor_cinco_meses = valor_cinco_meses + 1
    if x['tramo_mora_princip_06_A'] > 4: valor_cinco_meses = valor_cinco_meses + 1
    if x['num_meses_con_deuda_ult6'] == 0: valor_cinco_meses = -1

    return valor_cinco_meses

def ratio_mora_a_siete_meses(x):
    global valor_siete_meses
    valor_siete_meses = 0

    if x['tramo_mora_princip_01_A'] > 6: valor_siete_meses = valor_siete_meses + 1
    if x['tramo_mora_princip_02_A'] > 6: valor_siete_meses = valor_siete_meses + 1
    if x['tramo_mora_princip_03_A'] > 6: valor_siete_meses = valor_siete_meses + 1
    if x['tramo_mora_princip_04_A'] > 6: valor_siete_meses = valor_siete_meses + 1
    if x['tramo_mora_princip_05_A'] > 6: valor_siete_meses = valor_siete_meses + 1
    if x['tramo_mora_princip_06_A'] > 6: valor_siete_meses = valor_siete_meses + 1
    if x['num_meses_con_deuda_ult6'] == 0: valor_siete_meses = -1

    return valor_siete_meses

def generando_cartera_atrasada(row, columnas):
    global valor_sumado
    valor_sumado = 0
    for i in columnas:
        if pd.notna(row[i]):
            valor_sumado = valor_sumado + row[i]
    return valor_sumado

def generando_max_deuda(row, columnas):
    global maximo_valor
    maximo_valor = row[columnas[0]]
    for i in columnas:
        if pd.notna(row[i]):
            maximo_valor = row[i]
            break

    for i in columnas:
        if pd.notna(row[i]):
            if row[i] > maximo_valor:
                maximo_valor = row[i]
    return maximo_valor

def ratio_saldo_maximo_meses(x):
    if x['saldo'] > 0:
        if x['num_meses_con_deuda_ult6'] == 0:
            return -1
        else :
            if ((x['max_deuda'] > 0) | (x['max_deuda'] < 0)) & (pd.notna(x['max_deuda'])) :
                return x['saldo'] / x['max_deuda']

def generando_max_tramo_mora(row, columnas):
    global maximo_valor
    maximo_valor = row[columnas[0]]
    for i in columnas:
        if pd.notna(row[i]):
            maximo_valor = row[i]
            break

    for i in columnas:
        if pd.notna(row[i]):
            if row[i] > maximo_valor:
                maximo_valor = row[i]
    return maximo_valor
    
def ultima_cantidad_meses_mora(x):
    if x['num_meses_con_deuda_ult6'] == 0 : return -1
    elif (x['tramo_mora_princip_01_A'] > 0) & (x['tramo_mora_princip_01_A'] < 7) : return 1
    elif (x['tramo_mora_princip_02_A'] > 0) & (x['tramo_mora_princip_02_A'] < 7) : return 2
    elif (x['tramo_mora_princip_03_A'] > 0) & (x['tramo_mora_princip_03_A'] < 7) : return 3
    elif (x['tramo_mora_princip_04_A'] > 0) & (x['tramo_mora_princip_04_A'] < 7) : return 4
    elif (x['tramo_mora_princip_05_A'] > 0) & (x['tramo_mora_princip_05_A'] < 7) : return 5
    elif (x['tramo_mora_princip_06_A'] > 0) & (x['tramo_mora_princip_06_A'] < 7) : return 6
    else : return 0


def ratio_tramo_moroso(x):
    global valor_ratio_moroso
    valor_ratio_moroso = 0
    if x['saldo'] > 0 :
        if x['nuevo_tramo'] == 1 :
            valor_ratio_moroso = (x['atr_1_7_dias'] + x['atr_8_15_dias'] + x['atr_16_30_dias'] + x['atr_31_45_dias'] + x['atr_46_60_dias'] + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 2 :
            valor_ratio_moroso = (x['atr_8_15_dias'] + x['atr_16_30_dias'] + x['atr_31_45_dias'] + x['atr_46_60_dias'] + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 3 :
            valor_ratio_moroso = (x['atr_16_30_dias'] + x['atr_31_45_dias'] + x['atr_46_60_dias'] + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 4 :
            valor_ratio_moroso = (x['atr_31_45_dias'] + x['atr_46_60_dias'] + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 5 :
            valor_ratio_moroso = (x['atr_46_60_dias'] + x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 6 :
            valor_ratio_moroso = (x['atr_61_90_dias'] + x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 7 :
            valor_ratio_moroso = (x['atr_91_120_dias'] + x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 8 :
            valor_ratio_moroso = (x['atr_121_150_dias'] + x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 9 :
            valor_ratio_moroso = (x['atr_151_180_dias'] + x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 10 :
            valor_ratio_moroso = (x['atr_181_360_dias'] + x['atr_360_dias_a_mas']) / x['saldo']
        elif x['nuevo_tramo'] == 11:
            valor_ratio_moroso = (x['atr_360_dias_a_mas']) / x['saldo']
        else : valor_ratio_moroso = 0
    else : valor_ratio_moroso = 0

    return valor_ratio_moroso
```

```python
def nuevo_segmento(row):
    if row["segmento"] in ['B'] : return "B"
    elif row['segmento'] in ['D', 'F'] : return "D&F"
    elif row['segmento'] in ['A', 'C'] : return "A&C&E"
    elif row['segmento'] in ['E'] : return "A&C&E"
    
def deuda_ult_seis_meses_post(x):
    global acum
    acum = 0
    if (x['deuda_tot_01_d'] > 0): acum = acum + 1
    if (x['deuda_tot_02_d'] > 0): acum = acum + 1
    if (x['deuda_tot_03_d'] > 0): acum = acum + 1
    if (x['deuda_tot_04_d'] > 0): acum = acum + 1
    if (x['deuda_tot_05_d'] > 0): acum = acum + 1
    if (x['deuda_tot_06_d'] > 0): acum = acum + 1
    return acum
```

```python
#Agregando Nuevo Segmento
df_base_mod = df_base_mod.assign(segmento2=df_base_mod.apply(nuevo_segmento, axis=1))

print('Agrego segmento')
df_base_mod.head()
```
