# Carga Data Ventas de Clientes

```python
ventas = pd.read_csv('../dex/ventas.csv', delimiter=';', low_memory=False)

ventas.head()
```

```python
renamed_columns_ventas = {'month_id':'var_mes',
                          'customer_id':'ali_cliente_id',
                          'customer':'ali_cliente_nom',
                          'customer_dex_id':'dex_cliente_id',
                          'customer_dex':'dex_cliente_nom',
                          'cash_sale':'venta_cont',
                         'credit_sale':'venta_cred',
                          'amount_sale':'venta_tot',
                          'debt_on_time':'cobr_plazo',
                          'debt_delayed':'cobr_atraso',
                          'amount_debt':'cobr_tot',
                          'debt_period':'plazo_cobr',
                          'current_account_balance':'saldo_cta_corr',
                          'rejected_check_balance':'saldo_chq_rechaz',
                          'credit_days':'dias_cred'}

ventas = ventas.rename(columns=renamed_columns_ventas)

ventas.head()
```

```python
ventas = ventas[~ventas.isin(["#"]).any(axis=1)]
```

```python
ventas = ventas.drop_duplicates(subset = ventas.columns.to_list(), keep='first')

ventas
```

```python
ventas[['ali_cliente_id','dex_cliente_id']] = ventas[['ali_cliente_id','dex_cliente_id']].astype('int64')
```

```python
ventas[['venta_cont','venta_cred','venta_tot','cobr_plazo','cobr_atraso','cobr_tot','plazo_cobr','saldo_cta_corr','saldo_chq_rechaz','dias_cred']] = ventas[['venta_cont','venta_cred','venta_tot','cobr_plazo','cobr_atraso','cobr_tot','plazo_cobr','saldo_cta_corr','saldo_chq_rechaz','dias_cred']].fillna(0)

ventas
```

```python
for index, row in ventas.iterrows():
    if ((row['venta_cont'] < 0) and (row['venta_cred'] >= abs(row['venta_cont']))):
        ventas.at[index, 'venta_cred'] = row['venta_cred'] + row['venta_cont']
        ventas.at[index, 'venta_cont'] = 0
    elif ((row['venta_cred'] < 0) and (row['venta_cont'] >= abs(row['venta_cred']))):
        ventas.at[index, 'venta_cont'] = row['venta_cred'] + row['venta_cont']
        ventas.at[index, 'venta_cred'] = 0
```

```python
ventas['venta_tot'] = ventas['venta_cred'] + ventas['venta_cont']
```

```python
base_cosecha = pd.DataFrame()

for i in range(len(var) - 1, 11, -1):
    temporal = pd.pivot_table(
        ventas[(ventas.var_mes > var[i - 6]) & (ventas.var_mes <= var[i])],
        index="dex_cliente_id",
        aggfunc={
            "var_mes": "count",
            "venta_tot": ["mean", "std"],
            "venta_cred": ["mean", "max"],
            # "porcent_util_linea": "median",
            "dias_cred": "median",
        },
    )
    temporal.columns = [
        "plazo_venta_cred_6M",
        # "porcent_util_linea_6M",
        "cant_meses_6M",
        "venta_cred_max_6M",
        "venta_cred_mean_6M",
        "venta_tot_mean_6M",
        "venta_tot_std_6M"
    ]    
    temporal.reset_index(inplace=True)
    temporal["var_mes"] = var[i]
    base_cosecha = pd.concat([base_cosecha, temporal], axis=0)
    print(var[i],var[i-6])
    print(temporal.shape, var[i])

base_cosecha.head()
```

```python
base_cosecha.info()
```

```python
base_cosecha.loc[base_cosecha.venta_tot_std_6M.isnull(), "venta_tot_std_6M"] = 0
```

```python
base_cosecha2 = pd.DataFrame()

for i in range(len(var) - 1, 11, -1):
    temporal2 = pd.pivot_table(
        ventas[(ventas.var_mes > var[i - 6]) & (ventas.var_mes <= var[i])],
        index=["dex_cliente_id"],
        columns=["var_mes"],
        aggfunc={"venta_cred": "max"},
    )
    temporal2.columns = temporal2.columns.get_level_values(1)
    temporal2["max_mes_venta_cred"] = temporal2.idxmax(axis="columns")
    temporal2["dex_cliente_id"] = temporal2.index
    temporal2["var_mes"] = var[i]
    base_cosecha2 = pd.concat(
        [base_cosecha2, temporal2[["dex_cliente_id", "var_mes", "max_mes_venta_cred"]]],
        axis=0,
    )
    print(var[i],var[i-6])
    print(temporal2.shape, var[i])

base_cosecha2.head()
```

```python
def buscar_mes(df):
    for i in range(0, len(var)):
        if var[i] == df:
            return i
```

```python
base_cosecha2["dif_max_venta_cred"] = base_cosecha2["var_mes"].apply(
    buscar_mes
) - base_cosecha2["max_mes_venta_cred"].apply(buscar_mes)
base_cosecha2.index.rename("index", inplace=True)
base_cosecha2.head()
```
