```python
import pandas as pd
```

```python
data = pd.read_csv('../dex/clean/final.csv', low_memory=False)
data
```

```python
clientes = pd.read_csv('../dex/clientes.csv', delimiter=';', low_memory=False)
clientes
```

```python
df = data.merge(clientes, left_on=['customer_id'], right_on=['id'], how='left')
df
```

```python
df = df.drop(columns=['id', 'name'])
df
```

```python
antiguedad = pd.read_csv('../dex/antiguedad.csv', delimiter=';', low_memory=False)
antiguedad
```

```python
df = df.merge(antiguedad, left_on='customer_id', right_on='id', how='left')
df = df.drop(columns=['id'])

df
```

```python
df[df['creation_date'].isnull()]
```

```python
df = df[~(df['creation_date'].isnull())]
df
```

```python
df.info()
```

```python
df['creation_date'] = df['creation_date'].astype('datetime64')
df['antiguedad'] = ((pd.to_datetime(df['var_mes'], format="%Y%m") + pd.tseries.offsets.MonthEnd(1)) - df['creation_date']).dt.days
df
```

```python
df['antiguedad'].value_counts()
```

```python
df['type'].value_counts()
```

```python
df = df[df['type'].isin(['BODEGAS','PUESTO DE MERCADO','GASTRONOMÍA','PANIFICACIÓN','MAYORISTAS'])]
df
```

```python
df_bodegas = df[df['type'] == 'BODEGAS']
df_pdm = df[df['type'] == 'PUESTO DE MERCADO']
df_gastro = df[df['type'] == 'GASTRONOMÍA'] 
df_panif = df[df['type'] == 'PANIFICACIÓN']
df_mayorista = df[df['type'] == 'MAYORISTAS']
```

```python
df_bodegas.to_csv('../dex/bocas/bodegas.csv', index=False)
df_pdm.to_csv('../dex/bocas/pdm.csv', index=False)
df_gastro.to_csv('../dex/bocas/gastro.csv', index=False)
df_panif.to_csv('../dex/bocas/pani.csv', index=False)
df_mayorista.to_csv('../dex/bocas/mayorista.csv', index=False)
```
